#### 1. Write a "counter" class that can be incremented up to a specified maximum value, will print an error if an attempt is made to increment beyond that value, and allows reseting the counter.

```python
class Counter:
    def  __init__(self, maximum_number):
        self.a = 0
        self.maximum_number = maximum_number 
        
    def specified(self):
        if self.a < self.maximum_number:
            self.a +=1
        else:
            print("Error: Maximum value reached")
            
    def reset(self):
        self.a = 0 
                
    def __str__(self):
        return f"Counter Value: {self.a}"
    
    
```

```python
counter = Counter(10)
for i in range(12):
    counter.specified()
    print(counter)
    
counter.reset()
print(counter)
```

##### 7. Add a function to the object classes, including the base, that returns a list of 16 pairs of x and y points on the parameter of the object. 

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass

    @abstractmethod
    def perimeter_points(self):
        pass

class Rectangle(Shape):
    def __init__(self, length, width, x, y):
        self.__length = length
        self.__width = width
        self.__x = x
        self.__y = y

    def get_length(self):
        return self.__length

    def get_width(self):
        return self.__width

    def get_x(self):
        return self.__x

    def get_y(self):
        return self.__y

    def area(self):
        return self.__length * self.__width

    def perimeter(self):
        return 2 * (self.__length + self.__width)

    def perimeter_points(self):
        points = []
        for i in range(4):
            for j in range(4):
                if i % 2 == 0:
                    x = self.__x + (i // 2) * self.__length
                else:
                    x = self.__x + (i // 2) * self.__length
                if j % 2 == 0:
                    y = self.__y + (j // 2) * self.__width
                else:
                    y = self.__y + (j // 2) * self.__width
                points.append((x, y))
        return points

class Circle(Shape):
    def __init__(self, radius, x, y):
        self.__radius = radius
        self.__x = x
        self.__y = y

    def get_radius(self):
        return self.__radius

    def get_x(self):
        return self.__x

    def get_y(self):
        return self.__y

    def area(self):
        import math
        return math.pi * (self.__radius ** 2)

    def perimeter(self):
        import math
        return 2 * math.pi * self.__radius

    def perimeter_points(self):
        import math
        points = []
        for i in range(16):
            angle = 2 * math.pi * i / 16
            x = self.__x + self.__radius * math.cos(angle)
            y = self.__y + self.__radius * math.sin(angle)
            points.append((x, y))
        return points

```

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    def area(self):
        pass

   
    def perimeter(self):
        pass

    
    def perimeter_points(self):
        pass

   
    def is_inside(self, x, y):
        pass
 

    def overlaps(self, other):
        pass

class Rectangle(Shape):
    def __init__(self, length, width, x, y):
        self.__length = length
        self.__width = width
        self.__x = x
        self.__y = y

    def get_length(self):
        return self.__length

    def get_width(self):
        return self.__width

    def get_x(self):
        return self.__x

    def get_y(self):
        return self.__y

    def area(self):
        return self.__length * self.__width

    def perimeter(self):
        return 2 * (self.__length + self.__width)

    def perimeter_points(self):
        points = []
        for i in range(4):
            for j in range(4):
                if i % 2 == 0:
                    x = self.__x + (i // 2) * self.__length
                else:
                    x = self.__x + (i // 2) * self.__length
                if j % 2 == 0:
                    y = self.__y + (j // 2) * self.__width
                else:
                    y = self.__y + (j // 2) * self.__width
                points.append((x, y))
        return points

    def is_inside(self, x, y):
        return self.__x <= x <= self.__x + self.__length and self.__y <= y <= self.__y + self.__width

    def overlaps(self, other):
        if isinstance(other, Rectangle):
            return not (self.__x + self.__length < other.get_x() or
                        self.__x > other.get_x() + other.get_length() or
                        self.__y + self.__width < other.get_y() or
                        self.__y > other.get_y() + other.get_width())
        elif isinstance(other, Circle):
            closest_x = max(self.__x, min(other.get_x(), self.__x + self.__length))
            closest_y = max(self.__y, min(other.get_y(), self.__y + self.__width))
            distance_x = other.get_x() - closest_x
            distance_y = other.get_y() - closest_y
            return (distance_x ** 2 + distance_y ** 2) < (other.get_radius() ** 2)
        return False

class Circle(Shape):
    def __init__(self, radius, x, y):
        self.__radius = radius
        self.__x = x
        self.__y = y

    def get_radius(self):
        return self.__radius

    def get_x(self):
        return self.__x

    def get_y(self):
        return self.__y

    def area(self):
        import math
        return math.pi * (self.__radius ** 2)

    def perimeter(self):
        import math
        return 2 * math.pi * self.__radius

    def perimeter_points(self):
        import math
        points = []
        for i in range(16):
            angle = 2 * math.pi * i / 16
            x = self.__x + self.__radius * math.cos(angle)
            y = self.__y + self.__radius * math.sin(angle)
            points.append((x, y))
        return points

    def is_inside(self, x, y):
        import math
        return math.sqrt((x - self.__x) ** 2 + (y - self.__y) ** 2) <= self.__radius

    def overlaps(self, other):
        if isinstance(other, Circle):
            import math
            distance = math.sqrt((self.__x - other.get_x()) ** 2 + (self.__y - other.get_y()) ** 2)
            return distance < (self.__radius + other.get_radius())
        elif isinstance(other, Rectangle):
            closest_x = max(other.get_x(), min(self.__x, other.get_x() + other.get_length()))
            closest_y = max(other.get_y(), min(self.__y, other.get_y() + other.get_width()))
            distance_x = self.__x - closest_x
            distance_y = self.__y - closest_y
            return (distance_x ** 2 + distance_y ** 2) < (self.__radius ** 2)
        return False
```

```python
from abc import ABC, abstractmethod
import tkinter as tk

class Canvas:
    def __init__(self, width, height):
        self.root = tk.Tk()
        self.canvas = tk.Canvas(self.root, width=width, height=height)
        self.canvas.pack()

    def draw_rectangle(self, x1, y1, x2, y2, **kwargs):
        self.canvas.create_rectangle(x1, y1, x2, y2, **kwargs)

    def draw_circle(self, x, y, r, **kwargs):
        self.canvas.create_oval(x-r, y-r, x+r, y+r, **kwargs)

    def mainloop(self):
        self.root.mainloop()

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass

    @abstractmethod
    def perimeter_points(self):
        pass

    @abstractmethod
    def is_inside(self, x, y):
        pass

    @abstractmethod
    def overlaps(self, other):
        pass

    @abstractmethod
    def paint(self, canvas):
        pass

class Rectangle(Shape):
    def __init__(self, length, width, x, y):
        self.__length = length
        self.__width = width
        self.__x = x
        self.__y = y

    def get_length(self):
        return self.__length

    def get_width(self):
        return self.__width

    def get_x(self):
        return self.__x

    def get_y(self):
        return self.__y

    def area(self):
        return self.__length * self.__width

    def perimeter(self):
        return 2 * (self.__length + self.__width)

    def perimeter_points(self):
        points = []
        for i in range(4):
            for j in range(4):
                if i % 2 == 0:
                    x = self.__x + (i // 2) * self.__length
                else:
                    x = self.__x + (i // 2) * self.__length
                if j % 2 == 0:
                    y = self.__y + (j // 2) * self.__width
                else:
                    y = self.__y + (j // 2) * self.__width
                points.append((x, y))
        return points

    def is_inside(self, x, y):
        return self.__x <= x <= self.__x + self.__length and self.__y <= y <= self.__y + self.__width

    def overlaps(self, other):
        if isinstance(other, Rectangle):
            return not (self.__x + self.__length < other.get_x() or
                        self.__x > other.get_x() + other.get_length() or
                        self.__y + self.__width < other.get_y() or
                        self.__y > other.get_y() + other.get_width())
        elif isinstance(other, Circle):
            closest_x = max(self.__x, min(other.get_x(), self.__x + self.__length))
            closest_y = max(self.__y, min(other.get_y(), self.__y + self.__width))
            distance_x = other.get_x() - closest_x
            distance_y = other.get_y() - closest_y
            return (distance_x ** 2 + distance_y ** 2) < (other.get_radius() ** 2)
        return False

    def paint(self, canvas):
        canvas.draw_rectangle(self.__x, self.__y, self.__x + self.__length, self.__y + self.__width, outline="black", fill="blue")

class Circle(Shape):
    def __init__(self, radius, x, y):
        self.__radius = radius
        self.__x = x
        self.__y = y

    def get_radius(self):
        return self.__radius

    def get_x(self):
        return self.__x

    def get_y(self):
        return self.__y

    def area(self):
        import math
        return math.pi * (self.__radius ** 2)

    def perimeter(self):
        import math
        return 2 * math.pi * self.__radius

    def perimeter_points(self):
        import math
        points = []
        for i in range(16):
            angle = 2 * math.pi * i / 16
            x = self.__x + self.__radius * math.cos(angle)
            y = self.__y + self.__radius * math.sin(angle)
            points.append((x, y))
        return points

    def is_inside(self, x, y):
        import math
        return math.sqrt((x - self.__x) ** 2 + (y - self.__y) ** 2) <= self.__radius

    def overlaps(self, other):
        if isinstance(other, Circle):
            import math
            distance = math.sqrt((self.__x - other.get_x()) ** 2 + (self.__y - other.get_y()) ** 2)
            return distance < (self.__radius + other.get_radius())
        elif isinstance(other, Rectangle):
            closest_x = max(other.get_x(), min(self.__x, other.get_x() + other.get_length()))
            closest_y = max(other.get_y(), min(self.__y, other.get_y() + other.get_width()))
            distance_x = self.__x - closest_x
            distance_y = self.__y - closest_y
            return (distance_x ** 2 + distance_y ** 2) < (self.__radius ** 2)
        return False

    def paint(self, canvas):
        canvas.draw_circle(self.__x, self.__y, self.__radius, outline="black", fill="red")

class CompoundShape(Shape):
    def __init__(self):
        self.shapes = []

    def add_shape(self, shape):
        self.shapes.append(shape)

    def area(self):
        return sum(shape.area() for shape in self.shapes)

    def perimeter(self):
        return sum(shape.perimeter() for shape in self.shapes)

    def perimeter_points(self):
        points = []
        for shape in self.shapes:
            points.extend(shape.perimeter_points())
        return points

    def is_inside(self, x, y):
        return any(shape.is_inside(x, y) for shape in self.shapes)

    def overlaps(self, other):
        return any(shape.overlaps(other) for shape in self.shapes)

    def paint(self, canvas):
        for shape in self.shapes:
            shape.paint(canvas)

class RasterDrawing:
    def __init__(self):
        self.shapes = []

    def add_shape(self, shape):
        self.shapes.append(shape)

    def paint(self, canvas):
        for shape in self.shapes:
            shape.paint(canvas)

    def modify_shape(self, index, new_shape):
        if 0 <= index < len(self.shapes):
            self.shapes[index] = new_shape

```
